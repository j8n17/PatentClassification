# 2. 형상 기호 처리(보류)

```python
form_special = preprocess_dict['form_special']
form_form = preprocess_dict['form_form']

# 짧은 단어가 뒤로 가야함. 짧은 단어가 먼저 처리되면 안되기 때문.
form_special = sorted(form_special, key=len, reverse=True)
form_form = sorted(form_form, key=len, reverse=True)
```

```python
def show_form(form):
    texts = train_df['claims'].values
    indices = list(range(texts.shape[0]))
    random.shuffle(indices)

    for idx in indices:
        text = texts[idx]
        if type(text) != str:
            continue
        found_specials = [special for special in form_special if special in text]
        special_combinations = found_specials + [element for x in found_specials for element in (f"'{x}'", f"' {x}'", f"' {x} '", f"‘{x}’", f"“{x}”")]
        combinations = [element for x in special_combinations for element in (f'{x}{form}', f'{x} {form}', f'{x}  {form}')]
        found_specials_form = [combination for combination in combinations if combination in text]

        if found_specials_form:
            print(found_specials_form)
            print(text)
            break

show_form('형')
```

```python
def convert_form(text):
    if type(text) != str:
        return text
    found_specials = [special for special in form_special if special in text] # form_special 중에서 텍스트에 있는 것들
    special_combinations = found_specials + [element for x in found_specials for element in (f"'{x}'", f"' {x}'", f"' {x} '", f"‘{x}’", f"“{x}”")]
    combinations = [element for x in special_combinations for y in form_form for element in (f'{x}{y}', f'{x} {y}', f'{x}  {y}')]# 텍스트에서 찾은 형식 특수문자와 (형식) 또는 (띄어쓰기한 형식)의 조합
    found_specials_form = [combination for combination in combinations if combination in text] # 이 조합들중 텍스트에 있는 조합
    for pattern in found_specials_form:
        text = re.sub(re.escape(pattern), 'ꕕ', text) # 있는 조합들 ꕕ으로 먼저 변경 후, 괄호 삭제후, '모양'으로 변경
    return text

train_df['invention_title'] = train_df['invention_title'].apply(convert_form)
train_df['abstract'] = train_df['abstract'].apply(convert_form)
train_df['claims'] = train_df['claims'].apply(convert_form)

show_form('형')
```

# 3. 같은 역할을 하는 특수문자 통일

```python
convert_list = ['bracket', 'tilde', 'emphasis', 'dot', 'number', 'additional', 'normal', 'operator']

originals = [original for key in convert_list for original in preprocess_dict[key].keys()]
converts = [convert for key in convert_list for convert in preprocess_dict[key].values()]

def convert_brackets(text):
    if type(text)!=str:
        # claims 중 nan값이 있으로 에러처리
        return text
    for original, convert in zip(originals, converts):
        if convert == "NaN":
            convert = ''
        try:
            text = re.sub(re.escape(original), convert, text)
        except:
            text = re.sub(re.escape(original), re.escape(convert), text)
    return text
```

# 4. 괄호로 감싸져 있는 것들은 모두 삭제

```python
def remove_bracket(text):
    if type(text) != str:
        return text

    brackets = list(set(re.findall(r'\((?:[^()]|\([^()]*\))*\)', text))) # ['(dfdf)', '(errfed)', ...] 형식
    brackets = sorted(brackets, key=len, reverse=True)

    for remove_word in brackets:
        text = re.sub(re.escape(remove_word), "", text)

    return text

def bracket_length(text):
    brackets = re.findall(r'\((?:[^()]|\([^()]*\))*\)', text)
    return max(len(word) for word in brackets) if brackets else 0
```

```python
# test
find_special('(')
print('-'*10)
# train_df에 적용
train_df['invention_title'] = train_df['invention_title'].apply(remove_bracket)
train_df['abstract'] = train_df['abstract'].apply(remove_bracket)
train_df['claims'] = train_df['claims'].apply(remove_bracket)

# test
find_special('(')
```

# 18자 이상 텍스트 지우기, tif 텍스트 지우기

```python
def remove_long_words(text):
    if type(text) != str:
        return text
    # Define the regular expression pattern to match words with length >= 18
    # The pattern matches sequences of alphanumeric characters and symbols with length of 18 or more
    pattern = r'\b[\w\S]{18,}\b'

    # Replace the matched words with an empty string
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text
```

```python
train_df['invention_title'] = train_df['invention_title'].apply(remove_long_words)
train_df['abstract'] = train_df['abstract'].apply(remove_long_words)
train_df['claims'] = train_df['claims'].apply(remove_long_words)
```

# 남은 모든 특수문자 삭제

```python
def clean_text(text):
    if type(text) != str:
        return text
    # 영어, 숫자, 한글, 공백을 제외한 모든 문자를 삭제
    cleaned_text = re.sub(r'[^a-zA-Z0-9가-힣\s,.ꕕ⏼]', ' ', text)
    return cleaned_text

train_df['invention_title'] = train_df['invention_title'].apply(clean_text)
train_df['abstract'] = train_df['abstract'].apply(clean_text)
train_df['claims'] = train_df['claims'].apply(clean_text)
```

# ꕕ -> (모양), ⏼ -> (수식)

```python
def convert_form(text):
    if type(text) != str:
        return text
    text = re.sub('ꕕ', '(모양)', text)
    text = re.sub('⏼', '(수식)', text)
    return text

train_df['invention_title'] = train_df['invention_title'].apply(convert_form)
train_df['abstract'] = train_df['abstract'].apply(convert_form)
train_df['claims'] = train_df['claims'].apply(convert_form)

# test
find_special('(수식)')
```
